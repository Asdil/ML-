```python
# 转换数据类型
dollarizer = lambda x: float(x[1:-1]) # $2.3 -> 2.3
chipo.item_price = chipo.item_price.apply(dollarizer)
```

```python
# 'loc'是基于标签的索引  'A','B'....
# 'iloc'是基于位置的索引   1，2，3，4
chipo['item_price'].iloc[0]=np.nan
chipo
```

```python
chipo['new'] = chipo.quantity * chipo.item_price
chipo
```

```python
# 多条件筛选
chipo.loc[(chipo['quantity']==2) & (chipo['order_id'] < 3000), ['item_price', 'new']]
```

```python
# 行组合乘积
total = (chipo.quantity * chipo.item_price).sum()
print('总价为: ' ,total)
```

```python
# create a function
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0

# apply the function to the gender column and create a new column
users['gender_n'] = users['gender'].apply(gender_to_numeric)
```

```python
# 除法，可以不一一对应，自动按照一个值对应
users.groupby('occupation').gender_n.sum() / users.occupation.value_counts() * 100 
```

```python
# 拼接
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
data1 = pd.DataFrame(raw_data_1, columns = ['subject_id', 'first_name', 'last_name'])
data2 = pd.DataFrame(raw_data_2, columns = ['subject_id', 'first_name', 'last_name'])
data3 = pd.DataFrame(raw_data_3, columns = ['subject_id','test_id'])
```

```python
data1
```

```python
data2
```

```python
# 上下拼接
all_data = pd.concat([data1, data2], join='outer', ignore_index=True) # ignore_index=True,  axis=1,
all_data
```

```python
# 添加一列
pd.merge(all_data, data3, on='subject_id')
```
